```python
import pandas as pd
from collections import Counter
import re

def top_words(series, n=10):
    # Combine all non‑null strings, split on non‑letters, lower‑case
    words = re.findall(r'\b\w+\b', ' '.join(series.dropna()))
    return Counter(words).most_common(n)

print('Top 10 words in Wins:')
print(top_words(df['Share a win from the last week (what went well, something you enjoyed)']))

print('\nTop 10 words in Losses:')
print(top_words(df['Share a loss (something that was challenging or did not go well)']))

print('\nTop 10 words in Blockers:')
print(top_words(df['Share a blocker, if any (anything that stopped you from doing what you needed to do)']))
```

```python
import pandas as pd
from collections import Counter
import re

# a minimal stop‑word set – you can extend it later
stop = {
    'i', 'me', 'my', 'myself', 'we', 'our', 'ours', 'ourselves', 'you', 'your',
    'yours', 'yourself', 'yourselves', 'he', 'him', 'his', 'himself', 'she',
    'her', 'hers', 'herself', 'it', 'its', 'itself', 'they', 'them', 'their',
    'theirs', 'themselves', 'what', 'which', 'who', 'whom', 'this', 'that',
    'these', 'those', 'am', 'is', 'are', 'was', 'were', 'be', 'been', 'being',
    'have', 'has', 'had', 'having', 'do', 'does', 'did', 'doing', 'a', 'an',
    'the', 'and', 'but', 'if', 'or', 'because', 'as', 'until', 'while', 'of',
    'at', 'by', 'for', 'with', 'about', 'against', 'between', 'into', 'through',
    'during', 'before', 'after', 'above', 'below', 'to', 'from', 'up', 'down',
    'in', 'out', 'on', 'off', 'over', 'under', 'again', 'further', 'then',
    'once', 'here', 'there', 'when', 'where', 'why', 'how', 'all', 'any',
    'both', 'each', 'few', 'more', 'most', 'other', 'some', 'such', 'no',
    'nor', 'not', 'only', 'own', 'same', 'so', 'than', 'too', 'very', 's',
    't', 'can', 'will', 'just', 'don', 'should', 'now'
}

def top_meaningful_words(series, n=10):
    text = ' '.join(series.dropna())
    words = re.findall(r'\b\w+\b', text.lower())
    words = [w for w in words if w not in stop]
    return Counter(words).most_common(n)

print('Top 10 meaningful words in Wins:')
print(top_meaningful_words(df['Share a win from the last week (what went well, something you enjoyed)']))

print('\nTop 10 meaningful words in Losses:')
print(top_meaningful_words(df['Share a loss (something that was challenging or did not go well)']))

print('\nTop 10 meaningful words in Blockers:')
print(top_meaningful_words(df['Share a blocker, if any (anything that stopped you from doing what you needed to do)']))
```

```python
# Define a few themes and the words that signal them
themes = {
    "time_management": ["time", "busy", "schedule", "deadline"],
    "technical_issues": ["network", "wifi", "data", "software", "bug", "issue"],
    "course_work": ["course", "assignment", "lecture", "project", "study"],
    "personal_wellbeing": ["stress", "health", "family", "motivation", "energy"]
}

def assign_themes(series):
    # Start with an empty list for each row
    all_tags = []
    for text in series.dropna():
        tags = []
        lowered = text.lower()
        for theme, keywords in themes.items():
            if any(word in lowered for word in keywords):
                tags.append(theme)
        # If no keyword matches, call it “other”
        all_tags.append(tags or ["other"])
    return all_tags

# Apply to each column
wins_tags = assign_themes(df['Share a win from the last week (what went well, something you enjoyed)'])
losses_tags = assign_themes(df['Share a loss (something that was challenging or did not go well)'])
blockers_tags = assign_themes(df['Share a blocker, if any (anything that stopped you from doing what you needed to do)'])

# Helper to count tags
def count_tags(tag_list):
    flat = [item for sublist in tag_list for item in sublist]
    return Counter(flat).most_common()

print("Wins themes:")
print(count_tags(wins_tags))
print("\nLosses themes:")
print(count_tags(losses_tags))
print("\nBlockers themes:")
print(count_tags(blockers_tags))
```

```python
import pandas as pd
from collections import Counter
import string
from nltk.corpus import stopwords

# Make sure the stop‑word list is available
import nltk
nltk.download('stopwords')
stop = set(stopwords.words('english'))

def clean_word_counts(series, top_n=10):
    words = []
    for txt in series.dropna():
        if not isinstance(txt, str):
            continue
        txt = txt.lower().translate(str.maketrans('', '', string.punctuation))
        words.extend([w for w in txt.split() if w not in stop])
    return Counter(words).most_common(top_n)

# Get the cleaned counts
wins   = clean_word_counts(df['Share a win from the last week (what went well, something you enjoyed)'])
losses = clean_word_counts(df['Share a loss (something that was challenging or did not go well)'])
blockers = clean_word_counts(df['Share a blocker, if any (anything that stopped you from doing what you needed to do)'])

# Helper to turn the list of tuples into a DataFrame
def to_df(counter_list, name):
    df = pd.DataFrame(counter_list, columns=['word', 'count'])
    df['category'] = name
    return df

# Combine into one table (optional)
combined = pd.concat([
    to_df(wins,   'wins'),
    to_df(losses, 'losses'),
    to_df(blockers, 'blockers')
]).reset_index(drop=True)

# Show the table nicely
combined
```

```python
import string
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from nltk.sentiment import SentimentIntensityAnalyzer

# Make sure NLTK knows where its data lives
nltk.data.path.append(r'C:\Users\IkayR\AppData\Roaming\nltk_data')

stop = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
sia = SentimentIntensityAnalyzer()

def clean_text(text):
    """Lower‑case, strip punctuation, tokenise, remove stop‑words and lemmatise."""
    if not isinstance(text, str):
        return ''
    text = text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    tokens = nltk.word_tokenize(text)
    tokens = [lemmatizer.lemmatize(t) for t in tokens if t not in stop]
    return ' '.join(tokens)

def get_sentiment(series):
    """Return VADER compound sentiment score for each element in the series."""
    return series.apply(lambda x: sia.polarity_scores(x)['compound'])

# -------------------------------------------------
# Quick sanity checks (run after loading the CSV)
# -------------------------------------------------
print("DataFrame shape:", df.shape)                     # should not be (0, ...)
print("Columns present:", df.columns.tolist())          # confirm the free‑text column names
```

```python

# Load the data (using the same path as earlier)
import pandas as pd
csv_path = 'data/raw/Umuzi_XB1_Check_in.csv'
df = pd.read_csv(csv_path)

# Show the column headers so you can copy the exact names if needed
print(df.columns)

# Define the source columns (adjust only if the names differ)
win_col   = "Share a win from the last week (what went well, something you enjoyed)"
loss_col  = "Share a loss (something that was challenging or did not go well)"
block_col = "Share a blocker, if any (anything that stopped you from doing what you needed to do)"

# Simple cleaning function
def clean_text(text):
    if not isinstance(text, str):
        return ''
    return text.lower()

# Create the cleaned columns
df['win_clean']     = df[win_col].apply(clean_text)
df['loss_clean']    = df[loss_col].apply(clean_text)
df['blocker_clean'] = df[block_col].apply(clean_text)

# Quick sanity check
print(df[['win_clean', 'loss_clean', 'blocker_clean']].head())
```

```python
# -------------------------------------------------
# Imports for topic modelling
# -------------------------------------------------
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import numpy as np

# -------------------------------------------------
# Helper to train LDA and return topics
# -------------------------------------------------
def fit_lda(text_series, n_topics=5, n_top_words=5):
    """
    text_series : pandas Series of cleaned strings
    Returns a tuple (lda_model, vectorizer, doc_topics, top_words)
    """
    # TF‑IDF vectorisation
    vectorizer = TfidfVectorizer(
        max_df=0.95,          # ignore words that appear in >95 % of docs
        min_df=2,             # ignore words that appear in <2 docs
        stop_words='english'  # remove common stop‑words
    )
    X = vectorizer.fit_transform(text_series)

    # LDA model
    lda = LatentDirichletAllocation(
        n_components=n_topics,
        random_state=42,
        learning_method='online',
        max_iter=10,
        learning_offset=50.,
        doc_topic_prior=None,
        topic_word_prior=None
    )
    doc_topics = lda.fit_transform(X)   # shape = (n_docs, n_topics)

    # Extract top words per topic
    feature_names = vectorizer.get_feature_names_out()
    top_words = []
    for topic_idx, topic in enumerate(lda.components_):
        top = [feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]
        top_words.append(top)

    return lda, vectorizer, doc_topics, top_words

# -------------------------------------------------
# Apply to each column
# -------------------------------------------------
wins_lda, wins_vec, wins_doc, wins_topics = fit_lda(df['win_clean'])
losses_lda, losses_vec, losses_doc, losses_topics = fit_lda(df['loss_clean'])
blockers_lda, blockers_vec, blockers_doc, blockers_topics = fit_lda(df['blocker_clean'])

# -------------------------------------------------
# Helper to display topics
# -------------------------------------------------
def print_topics(top_words, prefix=""):
    for i, words in enumerate(top_words):
        print(f"{prefix}Topic {i+1}: {', '.join(words)}")

print("\nWins – top words per topic")
print_topics(wins_topics, prefix="W‑")

print("\nLosses – top words per topic")
print_topics(losses_topics, prefix="L‑")

print("\nBlockers – top words per topic")
print_topics(blockers_topics, prefix="B‑")

# -------------------------------------------------
# Assign the dominant topic to each row (optional)
# -------------------------------------------------
df['win_topic']     = np.argmax(wins_doc, axis=1) + 1   # +1 for 1‑based labelling
df['loss_topic']    = np.argmax(losses_doc, axis=1) + 1
df['blocker_topic'] = np.argmax(blockers_doc, axis=1) + 1

# -------------------------------------------------
# Plot topic frequencies (matplotlib)
# -------------------------------------------------
import matplotlib.pyplot as plt

def plot_topic_counts(series, title):
    counts = series.value_counts().sort_index()
    plt.figure(figsize=(6, 4))
    plt.bar(counts.index.astype(str), counts.values, color='steelblue')
    plt.title(title)
    plt.xlabel('Topic')
    plt.ylabel('Number of responses')
    plt.tight_layout()
    plt.show()

plot_topic_counts(df['win_topic'],     "Wins – Topic Distribution")
plot_topic_counts(df['loss_topic'],    "Losses – Topic Distribution")
plot_topic_counts(df['blocker_topic'], "Blockers – Topic Distribution")
```
